# Iris Flower Clasification

This project aim to classify the Iris flowers into their respective species using machine learning techniques.


```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df["species"] = iris.target
```

## Exploratory Data Analysis

The dataset contains total 150 samples with 4 numerical features related to sepal and petal measurements. The target variable 'species' has 3 classes and is evenly distributed across the dataset. No missing values were observed, making the dataset clean and suitable for classification

```python
x = df.drop("species", axis=1)
y = df["species"]
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42 )
```

## Data Preparation

The feature set and target variable were seperated to prepare the data modeling. The datastet was then split into training and testing sets, with 80% of the data used for training and 20% for testing. This helps evaluate the model unseen data.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=200)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

## Model Training and Evaluation

A Logistic Regression model was trained using the prepared training dataset. The model was evaluated on the test set to assess its performnce on unseen data. The result show that model is able to classify iris species with complete accuracy, indicating separation between classes based on the given features.
